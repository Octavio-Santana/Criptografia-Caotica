# Caos

    A idéia de um sistema ser determinístico e imprevisível, proposto por Poicaré com o problema dos três corpos, não foi aceito pela comunidade científica em sua epoca. Lorenz estudando problemas atmosféricos mostrou, utilizando cálculos computacionais, que pequenas variações nas condições iniciais poderiam acarretar grandes diferenças na evolução do sistema. 
    
    As equações de Lorentz na forma simplifica:
   \begin{equation}
   \frac{dx}{dt} = \sigma(y-x) \\
   \frac{dy}{dt} = x(\rho - z) - y\\
   \frac{dz}{dt} = xy - bz
   \end{equation}

```python
# Condições iniciais #
x0 = 1.0000
y0 = 1.01
z0 = 1.1

inic_1 = [x0, y0, z0]
inic_2 = [x0+0.0001, y0, z0]
print inic_1
print inic_2
```

```python
# Parâmetros #
s = 10.0
r = 28.0
b = 8.0/3.0
```

```python
# Equações diferenciais de Lorentz #
def f(u, time):
#    u[0] = x, u[1] = y e u[2] = z
    dx = s*( u[1] - u[0])
    dy = u[0]*( r - u[2] ) - u[1]
    dz = u[0]*u[1] - b*u[2]
    return [dx, dy, dz]
```

```python
# Espaço amostral #
t = np.linspace(0,500,10000)
```

```python
# Resolve as equações diferenciais para dois conjutos de 
# condições iniciais #
res_1 = odeint(f, inic_1, t)
res_2 = odeint(f, inic_2, t)
```

```python
xi, yi, zi = 0, 1, 2
```

```python
plt.plot(t,res_1[:,xi], lw=1, color='blue', label='x0=1.0000')
plt.plot(t,res_2[:,xi], lw=1, color='red', label='x0=1.0001')
plt.xlabel('time', fontsize=20)
plt.ylabel('x', fontsize=20)
plt.title('y0=1.01 e z0=1.1', fontsize=20)
plt.legend(loc='upper left')
plt.axis([300,350,-20,20])
plt.show()
```

```python
plt.plot(t,res_1[:,yi], lw=1, color='blue', label='x0=1.0000')
plt.plot(t,res_2[:,yi], lw=1, color='red', label='x0=1.0001')
plt.title('y0=1.01 e z0=1.1', fontsize=20)
plt.xlabel('time', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.legend(loc='upper left')
plt.axis([300,350,-30,30])
plt.show()
```

```python
plt.plot(t,res_1[:,zi], lw=1, color='blue', label='x0=1.0000')
plt.plot(t,res_2[:,zi], lw=1, color='red', label='x0=1.0001')
plt.title('y0=1.01 e z0=1.1', fontsize=20)
plt.xlabel('time', fontsize=20)
plt.ylabel('z', fontsize=20)
plt.legend(loc='upper left')
plt.axis([0,30,0,50])
plt.show()

# Percebe que no inicio as duas curvas são identicas, mas depois de algum tempo elas começas a se comportar 
# de maneira diferente
```

```python
plt.plot(t,res_1[:,zi], lw=1, color='blue', label='x0=1.0000')
plt.plot(t,res_2[:,zi], lw=1, color='red', label='x0=1.0001')
plt.title('y0=1.01 e z0=1.1', fontsize=20)
plt.xlabel('time', fontsize=20)
plt.ylabel('z', fontsize=20)
plt.legend(loc='upper left')
plt.axis([300,320,0,50])
plt.show()

# Aqui ilustra melhor, como as duas curvas estão totalmente diferente para um diferença na condição inicial
# de 0.0001.
```
