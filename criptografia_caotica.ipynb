```python
import numpy as np
from math import *
from scipy.integrate import *
```

```python
def key(x0, y0, z0, msg):
        
    # Condição inicial #
    inic = [x0, y0, z0]
    
    # Parâmentros #
    s = 10.0
    r = 28.0
    b = 8.0/3.0
    
    # Equações Diferenciais de Lorentz #
    def f(u, time):
        # u[0] = x, u[1] = y e u[2] = z #
        dx = s*( u[1] - u[0] )
        dy = u[0]*( r - u[2] ) - u[1]
        dz = u[0]*u[1] - b*u[2]
        return [dx, dy, dz]
    
    # Espaço amostral #
    t = np.linspace(0,500,10000)
    tf = 0
    dt = 1  
    # Caso seja verdade aumente o espaço amostral #
    while (2*len(t)/(3*10)) < len(msg):
        tf += 500
        dt *= 10
        t = np.linspace(0, 500+tf, 10000*dt)
    
    # A chave inicia nesta posição da lista t #
    n = 2*len(t)/(3*10)
    
    # Resolve a equação diferencial #
    res = odeint(f, inic, t)
    
    res_max = max(res[:,2])
    res_min = min(res[:,2])

    # Python 2.7 o chr() vai até 255, então essa condição é para evitar 
    # de sair do range do chr(). No python 3, esse if não é necessário (eu acho)
    while ( res_max + 127 ) >= 255:
        res[:,2] = res[:,2]/255        
    
    # Chave caótica #
    keys = [int(res[:,2][n+i*10]) for i in range(len(msg))]
    return keys    
```

```python
# Codifica a mensagem #
def enc(msg, key):
    s = ''
    for i, m in enumerate(msg):
        s += chr( abs(ord(m) + key[i]) )
    return s

# Decodifica a mensagem #
def dec(msg, key):
    s = ''
    for i, m in enumerate(msg):
        s += chr( abs(ord(m) - key[i]) )
    return s
```

```python
# Chave de entrada, que corresponde as condições iniciais das equações diferenciais #
numb = [1.0001, 1.01, 1.1]
```

```python
t = open('romeu.dat', 'r')
texto = ''
for f in t:
    texto += f
```

```python
print texto
```

```python
# Chave caótica #
chave = key(numb[0], numb[1], numb[2], texto)
```

```python
crip = enc(texto, chave)
print crip
```

```python
decrip = dec(crip, chave)
print decrip
```

```python
# Caso alguem tente descriptografar seu texto, mas ele cometa um mínimo erro nas chaves de entrada #
chave_fail = key(numb[0]+0.0001, numb[1], numb[2], texto)
decrip = dec(crip, chave_fail)
print decrip
```
